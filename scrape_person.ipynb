```python

from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# download the chromedriver.exe from https://chromedriver.storage.googleapis.com/index.html?path=106.0.5249.21/
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
options.binary_location = "/usr/bin/google-chrome"
driver = webdriver.Chrome(options=options)


driver.get("https://www.linkedin.com/home")
time.sleep(1)
driver.find_element(by=By.XPATH, value="//input[@type='text']").send_keys("yukendiran.k@ideas2it.com")
driver.find_element(by=By.XPATH, value="//input[@type='password']").send_keys("Yuki@2001")
time.sleep(3)
driver.find_element(by=By.XPATH, value="//button[@data-id='sign-in-form__submit-btn']").click()
cookies_dict = {}
for cookie in driver.get_cookies():
    cookies_dict[cookie['name']] = cookie['value']
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36',
                        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
                        'accept-encoding': 'gzip, deflate, br',
                        'accept-language': 'en-US,en;q=0.9',
                        'upgrade-insecure-requests': '1',
                        'scheme': 'https'}
```
